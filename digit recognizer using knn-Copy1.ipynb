```python
main_df=df.copy()
df.drop(['label'],axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
hii=MinMaxScaler()
df1=hii.fit_transform(df)
df1=pd.DataFrame(df1,index=df.index,columns=df.columns)
```

```python
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
hlo=StandardScaler()
df2=hlo.fit_transform(df1)
df2=pd.DataFrame(df2,index=df1.index,columns=df1.columns)
```

```python
df2
```

```python
# now normalization
from sklearn.preprocessing import Normalizer
akh=Normalizer()
df3=akh.fit_transform(df2)
df3=pd.DataFrame(df3,index=df2.index,columns=df2.columns)
```

```python
df3
```

```python
df4=pd.concat([main_df['label'],df3],axis=1)
```

```python
df4.head()
```

```python
Y_train=df4['label']
X_train=df4.drop(['label'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
Y_train= enc.fit_transform(Y_train)
```

```python
Y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,Y_train)
```
